```python
import pandas as pd
import os
```

```python
# 데이터를 저장할 위치로 이동
% cd /content/drive/MyDrive/semi-projet/data
```

```python
# 합칠 파일들 리스트 확인
file_list = os.listdir("./")
file_list
```

```python
!pwd
```

```python
# Dacon 데이터 로드 (총 45654개)
PATH ='/content/drive/MyDrive/semi-projet/data'
dacon_df = pd.read_csv(PATH + '/news_topic/train_data.csv')
dacon_df.info()
```

```python
dacon_df.head()
```

```python
# Dacon 데이터와 크롤링 데이터를 concat
# => Total 140773개의 데이터
no_file_list = ['news_topic', 'train.csv', 'temp']
for f in file_list:
    if f not in no_file_list:
        crawling_df = pd.read_csv(f)
        print(f'{f}:', len(crawling_df))
        dacon_df = pd.concat([dacon_df, crawling_df])
dacon_df.info()
```

```python
# merge 후 데이터 확인
dacon_df.head()
```

```python
# 불필요한 column 제거
dacon_df.drop(['index', 'Unnamed: 0'], axis=1, inplace=True)
dacon_df.head()
```

```python
# Data 셔플
# 토픽별로 크롤링한 데이터를 그대로 합쳤으니 한번 셔플링해줍니다
df_shuffled=dacon_df.sample(frac=1).reset_index(drop=True)
df_shuffled
```

```python
# PATH에 train.csv로 저장
df_shuffled.to_csv(PATH + '/train.csv')
```
