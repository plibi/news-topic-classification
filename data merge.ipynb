```python
# 합칠 파일들 리스트 확인
file_list = os.listdir("./")
file_list
```

```python
PATH ='/content/drive/MyDrive/semi-projet/data'
dacon_df = pd.read_csv(PATH + '/news_topic/train_data.csv')
dacon_df.info()
```

```python
dacon_df.head()
```

```python
no_file_list = ['news_topic', 'train.csv', 'temp']
for f in file_list:
    if f not in no_file_list:
        crawling_df = pd.read_csv(f)
        print(f'{f}:', len(crawling_df))
        dacon_df = pd.concat([dacon_df, crawling_df])
dacon_df.info()
```

```python
dacon_df.head()
```

```python
dacon_df.drop(['index', 'Unnamed: 0'], axis=1, inplace=True)
dacon_df.head()
```

```python
# Data 셔플
df_shuffled=dacon_df.sample(frac=1).reset_index(drop=True)
df_shuffled
```

```python
# PATH에 train.csv로 저장
df_shuffled.to_csv(PATH + '/train.csv')
```
